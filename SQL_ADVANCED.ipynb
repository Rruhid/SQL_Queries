## Write a query that gives an overview of how many films have replacements costs in the following cost ranges
## Напишите запрос, который дает обзор того, сколько фильмов имеют стоимость  в следующем диапазоне


```python
#low: 9.99 - 19.99
#medium: 20.00 - 24.99
#high: 25.00 - 29.99

%sql SELECT\
CASE WHEN replacement_cost BETWEEN 9.99 AND 19.99 THEN 'Low'\
WHEN replacement_cost BETWEEN 20.00 AND 24.99 THEN 'Medium'\
WHEN replacement_cost BETWEEN 25.00 AND 29.99 THEN 'High'\
END AS Level, COUNT(*) AS Count\
FROM film GROUP BY Level
```

# Write a query that returns the customer's name, country and how many payments they have?
# Вернем запрос, возвращающий имя клиента, страну и сколько у него платежей?

```python
%sql SELECT * FROM\
(SELECT first_name, last_name, cc.country, COUNT(*) AS Count,\
RANK() OVER(PARTITION BY cc.country ORDER BY COUNT(*) DESC) AS r_k\
FROM customer c INNER JOIN country cc\
ON c.address_id =cc.country_id INNER JOIN payment p ON c.customer_id = p.customer_id GROUP BY\
first_name, last_name, cc.country) a\
WHERE r_k IN (1,3)
```

```python
%sql SELECT * FROM\
(SELECT first_name, last_name, cc.country, COUNT(*) AS Count,\
FIRST_VALUE(COUNT(*)) OVER(PARTITION BY cc.country ORDER BY COUNT(*) DESC) AS r_k\
FROM customer c INNER JOIN country cc\
ON c.address_id =cc.country_id INNER JOIN payment p ON c.customer_id = p.customer_id GROUP BY\
first_name, last_name, cc.country) a
```

# Let's count the percentage difference 

# lEAD AND LAG FUNCTIONS  

```python
%sql SELECT  DATE(payment_date), SUM(amount) AS Sum,\
LAG(SUM(amount)) OVER(ORDER BY DATE(payment_date)) AS Prev_Month,\
SUM(amount) -  LAG(SUM(amount)) OVER(ORDER BY DATE(payment_date)) AS Difference_1,\
ROUND((SUM(amount) -  LAG(SUM(amount)) OVER(ORDER BY DATE(payment_date)))  / LAG(SUM(amount))\
OVER(ORDER BY DATE(payment_date)) *100, 1) AS Perc_dif\
FROM payment GROUP BY DATE(payment_date)
```
